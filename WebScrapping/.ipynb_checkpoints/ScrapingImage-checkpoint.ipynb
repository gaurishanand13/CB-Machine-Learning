```python
from urllib.request import urlopen #Libriary to make a request on the website to get data
image_url = "https://assets.passiton.com/quotes/quote_artwork/5572/original/20200512_tuesday_quote.jpg?1588967401"

data_fetched = urlopen(image_url)
print(type(data_fetched))
print(data_fetched)
```

```python
data = data_fetched.read()
print(type(data))
print(data) #The data printed is actually an image only converted into some binary code. 
#Therefore to get this image back in the original format which we were able to see on the web. For that we need
#to save the image in a file (i.e binary code is interpreted by the file). Then it would be open like that only
```

```python
#The wb indicates that the file is opened for writing in binary mode. When writing in binary mode, 
#Python makes no changes to data and it is written as it is to the file. It is basically used to store images(jpg,jpeg,etc)
#exe files.
with open("myFirstImage.jpg",'wb') as f:
    f.write(data)
    
#Now you will be able to see an image named as 'myFirstImage.jpg' saved in the folder which will be the same image
#which was in the url. Therefore we were successfully able to download and scrap the image
```

# Scraping a website having multiple images and download all the images

```python
url = "https://www.passiton.com/inspirational-quotes?page=2"
data_fetched = urlopen(url)
data = data_fetched.read()

#Now we would convert it into an beautiful soup object, so that we can filter our required data
from bs4 import BeautifulSoup as soup
myDataSoup = soup(data,'html.parser')

myDataSoup = myDataSoup.findAll('div',{
    'class':'row',
    'id' : 'all_quotes'
})
print(len(myDataSoup))
```

```python
myDataSoup =  myDataSoup[0].findAll('div',{})
print(len(myDataSoup))
```

```python
myList = []
for x in myDataSoup:
    y = x.find('img')
    #Now in this particular tag , if we want to access some attribut, we use [] for that
    myList.append(y['src'])

"\n".join(myList)
```

```python
for index,s in enumerate(myList):
    data_fetched = urlopen(image_url)
    data = data_fetched.read()
    with open("image{}.jpg".format(index),'wb') as f:
        f.write(data)
```
