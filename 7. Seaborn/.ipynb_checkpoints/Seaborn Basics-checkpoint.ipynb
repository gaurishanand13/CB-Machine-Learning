```python
#Let's visualize this dataset
tips = sns.load_dataset('tips')
```

#### Scatter Plot

```python
sns.scatterplot(x = 'total_bill',y='tip',data=tips) #There we should always have the datafram object for the data which
#needs to be plotted using seaborn
```

#### Barplot

```python
#let's plot the bar graph of let's say size
sns.barplot(x='size',y='total_bill',data=tips)

#Here plot will be plotted where (x = size) and at y for that that particular
#x , we will be having (e.g = at x =1 (i.e size =1) , Y will be the mean of all the total_bill where size=1)
```

```python
#But if we want, for each x(size) . Instead of plotting mean on Y axis, Plot should be done for standard deviation
sns.barplot(x='size',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
#Therefore the below graph shows the avg comparison between total_bills of Female vs Male
```

#### Countplot - Counting the occurance of certain values i.e Basically Histogram

```python
sns.countplot(x='sex',data=tips)
#Therefore in data in total we have 150+Males and only 80+Females
```

```python
sns.countplot(x='time',data=tips)
#Therefore clearly , from this plot we can conclude that mostly people arrive for dinner in the restaurant
```

```python
sns.countplot(x='total_bill',data=tips) #Therefore it is not exactly histogram, it plots for every unique x .
#But histogram works in range i.e it plot counts of values in certain range
```

### Distplot (or Histogram) and Kdeplot

```python
### we can even remove this line, which approximates this histogram
sns.distplot(tips.get('total_bill'),kde=False)
```

```python
#We can even increase or decrease these bins(or in other wards) like this
sns.distplot(tips.get('total_bill'),kde=False,bins=25)
```

```python
sns.kdeplot(tips.get('total_bill'))
```

#### BOXPLOT 

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

# CategoriaL Plots

### Box Plot & Violin Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
"""
Here let's anayalyse for x='Thurs'. Note at thursday , from the box we can observe that most of our bills will be b/w
20 and 12.
And then the number of total_bills between (20,30) and (12,8) is very less than the box. 
Clearly 8 is the minimum total bill on thursday and we have very few (around 4-5) total_bills on thursday which are 
more than  30 which is clear from this plot
"""

"""
Note this middle line in the box plot represents the average value of total_bills on each day. Therefore we can 
clearly see that on sunday , average value of total_bills is much higher as compared to other days.
"""

"""Therefore the above explaination made for the box plot clearly can be seen using violen plot"""
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

### Box Plot & Violin Plot for total_bills (Male vs Female)

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex') #It will divide our data in 2 parts(one for male and other
#for female and then plot it)

"""
Therefore from the graphs , we can conclude that average total_bills paid by male on each day is higher as compared
to females
"""
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

### Comparing 2 properties of the dataset using jointplot

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#Therefore clearly we can see that there is some +ve co-relation between x(i.e total_bill) and y(i.e tip) as
#most of the x increases with increase in y

#Also along with the their values plotted, we also have a frequency histogram of x and y at top and right .
#Histogram on top shows the frequency of x in that interval and similarly there is a histogram on right too.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')

#Therefore by default points are of scatter type. But we can also of other types like hex type
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')

#Hexgram where it is very dark shows that there are many points, and where it is light hexagram shows very less points
```

### Comparing all properties of the dataset using pairplot

```python
# Pair plot basically plots graph between each property of the dataset which deals with numbers
sns.pairplot(tips)

#Therefore let's analyse the first column
#1) Here total_bill vs total_bill is plotted i.e basically a historgam of it
#2) Here tips vs total_bill is plotted. Clearly we can see that tip and total_bill are +vely co-related(though 
#vy very less amount) to each other as increase in tips is almost leading to increase in total_bill
#3) In third , size vs tota_bill is plotted
```

```python
sns.pairplot(tips,hue='sex')
```

```python
# We can confirm the co-relation of each pair(which deals with numbers in the data set) whose graph i have drawn 
# above. Pandas has a special feature which calculates the co-relation between each pair in the dataset(DataFrame Object)
#which deals with the numbers

tips_corr = tips.corr()
print(type(tips_corr))
tips_corr
```

```python
# Therefore since, each pair is +vely co-related to each other. Therefore increase in one value of the pair will also
# lead to increase in the other value of pair. Also decrease in one value of the pair will also lead to dec in the 
#other.

sns.heatmap(tips_corr) #Note - Heatmap is the only plot in which 1st column and 1st row contains the name of the 
#properties which are linked. Clearly more is the value of the co-relation (or more is the value of the box), brighter 
#will be the colour of that Box
```

```python
sns.heatmap(tips_corr,annot=True)
```

# Data Visualization using Heatmap

```python
flights_Dataset = sns.load_dataset('flights')
flights_Dataset.head(n=15)
```

```python
#Therefore this dataset shows , how many people travelled by flight in yth month of an xth year.
#Now let's somewhat modify the data order to plot it using heatmap
flight_pivot = flights_Dataset.pivot_table(index = 'month',columns='year',values='passengers')
```

```python
flight_pivot
```

```python
#Therefore now we can easily visualize this using Heatmap
sns.heatmap(flight_pivot)

#Clearly more brighter colours appear as we move from 1949 to 1960 i.e more people travelled by flights with time.
#Also clearly in the month of july and august in the year 1960 is the most when the people travelled by flights.
```
