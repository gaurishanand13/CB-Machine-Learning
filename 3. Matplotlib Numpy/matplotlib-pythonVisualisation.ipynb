```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot on GRAPH

```python
#For plotting any graph like this, we need to provide some of the x values and corresponding y values for it. e.g - 
#for a straight line x and y values will look like this -

x = np.arange(10)
y = 2*x + 3

print(x)
print(y)
```

```python
plt.plot(x,y) #First parameter gets the values of x and 2nd parameter is the corresponding values of y for that x
plt.show()
```

```python
#We can even provide some extra paramteres like colour of the line represent f(x), title to the graph.
plt.plot(x,y,color = 'red')
plt.title('Straight Line')
plt.show()
```

```python
#We can also label what x represents and what y represents like this -> 
plt.plot(x,y,color = 'red')
plt.xlabel('x -->')
plt.ylabel('2x+3  -->')
plt.title('Straight Line')
plt.show()
```

```python
#We can also represent what specifically f(x) represents by adding a label to y = f(x)
plt.plot(x,y,color = 'red',label = "Value of f(x)")
plt.legend()
plt.xlabel('x -->')
plt.ylabel('2x+3  -->')
plt.title('Straight Line')
plt.show()
```

```python
#We can also represent the points marked in the graph of f(x)
plt.plot(x,y,color = 'red',label = "Value of f(x)",marker = '*')#Each point in the graph of f(x) will be represented as *
plt.legend()
plt.xlabel('x -->')
plt.ylabel('2x+3  -->')
plt.title('Straight Line')
plt.show()
```

```python
#We can also represents f(x) by a dashed line. Note,if we represent f(x) by a dashed line -> then we can't represent
#the point plotted on the line. 
plt.plot(x,y,color = 'red',label = "Value of f(x)",linestyle='dashed')
plt.title('Straight Line')
plt.show()
```

#### Scatter Plots

```python
#Instead of representing the whole f(x), if we just want to plot points of (x,y) provided. Then that can also be done
plt.scatter(x,y,color='red')
plt.show()
```

```python
#Instead of representing the whole f(x), if we just want to plot points of (x,y) provided. Then that can also be done
plt.scatter(x,y,color='red',label = "Value of f(x)",marker='*')
plt.legend()
plt.show()
```

### Representing 2 functions in the same graph - 

```python
x = np.arange(10)
y1 = 2*x + 3
y2 = x**2

print('x = ',x)
print('y1 = ',y1)
print('y2 = ',y2)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('Quantity of fruit vs Price')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.show()
```

```python
#Clearly we can't see which graph represents which function. Therefore we should give label to each f(x) so that we 
#distinguish
plt.plot(x,y1,label = 'Apple')
plt.plot(x,y2,label = 'Kiwi')
plt.legend()
plt.title('Quantity of fruit vs Price')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.show()
```

```python
#We can even change the colour of each line representing it
plt.plot(x,y1,color = 'red',label = 'Apple',marker = '*')
plt.plot(x,y2,color='purple',label = 'Kiwi',marker = 'o')
plt.legend()
plt.title('Quantity of fruit vs Price')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.show()
```

```python
#We can also change the size of the graph outputed plotted using the property figure
plt.figure(figsize=(10,10))

plt.plot(x,y1,color = 'red',label = 'Apple',marker = '*')
plt.plot(x,y2,color='purple',label = 'Kiwi',marker = 'o')
plt.legend()
plt.title('Quantity of fruit vs Price')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.show()
```

## BAR GRAPH

```python
#Therefore to plot a bar graph -> we need to have the value of x and correspondingly value of height for that x
x = [0,1,2]
height = [10,20,30]


plt.bar(x,height)
plt.show()
#By default as we can see that -> width of each bar is 0.8, where middle point of each bar is the corresponding 
#x provided
```

```python
#plt.bar?
plt.bar(x,height,width = 1)
plt.show()
```

```python
#Giving label to the Bar graph
plt.bar(x,height,width=0.2,label = 'Height of Students')
plt.legend()
plt.show()
```

```python
#We can also give which student's height it represents by providing the tick_label
plt.bar(x,height,width=0.2,label = 'Height of Students',tick_label=['Radhika','Neetu','Gaurish'])
plt.legend()
plt.show()
```

### Representing 2 bar graphs in one plot

```python
#We can clearly see that 2 bar graphs are overlapping since they have the same x
x1 = np.array([0,1,2])
height1 = np.array([5,10,15]) 
x2 = np.array([0,1,2])
height2 = np.array([5,7,10])
plt.bar(x1,height1)
plt.bar(x2,height2)
plt.show()
```

```python
#Clearly we can above graph is not clear as both the bars are overlapping. Therefore try to reduce the width and adjust
#x coordintaes of one of the bargraphs , so that they come adjacent to each other on the same x
x1 = np.array([0,1,2])
height1 = np.array([5,10,15]) 
x2 = np.array([0,1,2]) + 0.4
height2 = np.array([5,7,10])
plt.bar(x1,height1,width = 0.4,label = 'Height')
plt.bar(x2,height2,width = 0.4,label = 'Weight')
plt.legend()
plt.show()
```

```python
#Changing the colour of the bar graph / giving x and y labels can also be done here 
plt.bar(x1,height1,width = 0.4,label = 'Height',color = 'red')
plt.bar(x2,height2,width = 0.4,label = 'Weight',color = 'yellow')
plt.legend()
plt.xlabel('Height/Weight->')
plt.ylabel('Measure of Quantitiy->')
plt.show()
```

```python
#We can also give the space , that should be there after and before graphs ends in both x and y direction. It will 
#stretch till that value using xlim and ylim
plt.bar(x1,height1,width = 0.4,color = 'red')
plt.bar(x2,height2,width = 0.4,color = 'yellow')
plt.ylim(-5,20)
plt.xlim(-5,20)
plt.show()
```

## PieCharts

```python
subjects = ["Maths","Chem","Physics","English"]

weightage = [20,15,17,7] #It is not necessary that weightage sum should be 100. It will calculate percentage accroding
#to dataa
```

```python
plt.pie(weightage,labels = subjects)
plt.show()
```

## Normal Distribution -> Plotting a Histogram to show Normal Distribution

```python
Xsns = np.random.randn(100) ##It will generate 100 random numbers from a standard normal distribution i.e
#Most of the numbers will be between -1 and 1 and their mean will be 0 along with standard deviation = 1
print(Xsns)
```

```python
#If , we will round off these and observe these values, most of them are between -1 and 1. 
print(np.round(Xsns))
```

```python
#Let's plot the histogram of above to see whether it a standard normal distribution or not
plt.style.use('seaborn')
plt.hist(Xsns)
plt.show()
```

```python
#Clearly above takes the form of a normal distribution curve, not let's shift the mean of above graph to some other 
#value. Let's make mean(u) = 60 with sigma still as 1
u = 60  #(u = MEAN)
plt.hist(Xsns + u)
plt.show()

#Let's also observe the values plotted
print(np.round(Xsns+u)) ##Most of the values are around 60
```

```python
#Now let's increase the sigma = 15 i.e most values of x will be between (u-sigma) and (u+sigma). 
#Since sigma is increasing , therefore the graph should become more flatter in between
sigma = 20
u = 60
plt.hist(Xsns*sigma + u)
plt.show()

#THerefore , now value has gone even till less than 20 and even more than 100 i.e values are more spread as compared to above
print(np.round(Xsns*sigma+u)) ##Values are more spread
```

```python
#Note as seen above, we only have 8 bars, though we have data of 100s of data, if we want, we can increase the 
#frequency of bars i.e more clear overview of the distribution can be seen
plt.hist(Xsns*5 + 60,25)
plt.show()
```

```python
vals =np.round(Xsns*5 + 60)
z = np.unique(vals,return_counts=True) ##Returns the count of each x , it comes in the data i.e frequency of each x
print(z)
#51 comes 1 time
#52 comes 4 times
#53 comes 5 times
#.... Clearly most of the value of are mean (60-sigma) and (60+sigma) where sigma is 10
```

```python
##We can also show it more clearly that x occurs mainly in between the range (mean-sigma),(mean+sigma) --> Hence is 
##the use. Eg - 
values = np.random.randn(100)
```

```python
Xns = values*5 + 60 #Here mean is 60 and std deviation is 5
#For now let's assume that all values of x comes only have height as 1 --->
heights= np.ones(Xns.shape)
plt.scatter(Xns,heights)
plt.show() ##Clearly more values of x comes near the mean as we have dense there
```

## Multivariat Normal Distribution
<img src='multivariantExamplePicture.jpeg'>

```python
#This is how a plot looks in 3d , Clearly when sigma xy is 0 , x and y are not correlated to each other . Eg. see down
mean = np.array([0.0,0.0]) ##First paramter is the mean in x direction and the other is mean in y direction

covariance = np.array([
    [1,0],
    [0,1]
]) #Sigma xy = 0 i.e they are in no way co-related to each other

values = np.random.multivariate_normal(mean,covariance,500) #It will get us 500 values, following such distribution
print(values.shape)


plt.scatter(values[:,0],values[:,1])
plt.show() #Clearly sigma x and y and not co-related to each other in any way . Conider the case -> increase in x is
#leading to increase and decrease in y both. Therefore not related to other. In 3d it will look like the above figure
#in 1st
```

```python
#Now consider the case when we make sigma xy as +ve , then increase in should also lead to increase in y in most cases
mean = np.array([0.0,0.0])

covariance = np.array([
    [1,0.5],
    [0.5,1]
])

values = np.random.multivariate_normal(mean,covariance,500)
print(values.shape)


plt.scatter(values[:,0],values[:,1])
plt.show() #Now clearly , the graph is somewhat tilted. Therefore some +ve corelation. Let's increase sigma xy now
#to check how steeply can y increase when x increases
```

```python
mean = np.array([0.0,0.0])

covariance = np.array([
    [1,0.8],
    [0.8,1]
])

values = np.random.multivariate_normal(mean,covariance,500)
print(values.shape)


plt.scatter(values[:,0],values[:,1])
plt.show()
```

```python
mean = np.array([0.0,0.0])

covariance = np.array([
    [1,1],
    [1,1]
])

values = np.random.multivariate_normal(mean,covariance,500)
print(values.shape)


plt.scatter(values[:,0],values[:,1])
plt.show()#Clearly on making sigma xy as 1, x will be linearly dependeny on y i.e increase in x will lead to sure 
#increase in y 
```

```python
###Plotting 2 multivariate distribution on a single graph
mean1 = np.array([0.0,0.0])
mean2 = np.array([6.0,5.0])

covariance1 = np.array([
    [1.3,-0.8],
    [-0.8,1.3]
])
covariance2 = np.array([
    [1.3,0.2],
    [0.2,1.3]
])
values1 = np.random.multivariate_normal(mean1,covariance1,500)
values2 = np.random.multivariate_normal(mean2,covariance2,500)

plt.scatter(values1[:,0],values1[:,1])##Since negatively related , therefore increase in x leads to increase in y
plt.scatter(values2[:,0],values2[:,1]) ##Clearly it will be more close to circle as sigma x and y is very less almost 0
plt.show()
```
