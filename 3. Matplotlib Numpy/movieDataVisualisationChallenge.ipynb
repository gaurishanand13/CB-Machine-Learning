```python
#A real life example where we will be plotting a graph
#We will be plotting a Histogram where x =  length of the movie's title i.e bar's height denotes --> number
#of movies with length as x
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('movie_metadata.csv')
df.head(n=10)
```

```python
#First let's check which all columns are there in this ->
df.columns
```

```python
#Now since we want to work on movie_title column , we will extract it-->
myList = []
for i in range(len(df)):
    myList.append(df.iloc[i]['movie_title'])
```

```python
#Instead of making a for loop like above , we can get one column directly in pandas from 'data frame' object like this ->
myList = df.get('movie_title')
print(type(myList))  ##It will be a list but of pandas.series type list, therefore i type casted it to a type of 
myList = list(myList)## ---> python list here
print(type(myList))
```

```python
#Now i will store , the number of title with length x in a dictionary ->
myDictionary = {}
for title in myList:
    tempLength = len(title)
    if myDictionary.get(tempLength) is None:
        myDictionary[tempLength] = 1
    else:
        myDictionary[tempLength] = myDictionary[tempLength] + 1
    
print(myDictionary)
```

```python
#Now let's first plot a scattered plot for above
x = np.array(list(myDictionary.keys()))
y = np.array(list(myDictionary.values()))
print(type(x),type(y))
```

```python
plt.scatter(x,y)
plt.xlabel('Length of the movie title -->')
plt.ylabel('No of movies-->')
plt.title('Movie Data Visualisation Problem')
plt.show()
```
