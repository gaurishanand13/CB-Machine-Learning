### Locally weighted Regression Algo

```python
## Fist Finding the weight matrix i.e W(mXm) Matrix

def getW(query_point ,X ,tau): 
    m = X.shape[0] #Total Examples in W
    
    #Now forming a numpy Idendity matrix instead of numpy array
    W = np.mat(np.eye(m)) #Formed a numpy Identity matrix of size m where W1, W2 .. will be stored
    
    for i in range(m):
        #Note distance between 2 points is (X[i] - query_point)^2. Which is just a dot product of vectors and can be
        #computed easily the formula (X1 - X2)^2 = transpose(X1-X2)*(X1-X2). Because it works in the same manner
        
        #For explaination see image below
        numerator = np.sum((X[i] - query_point)**2)
        denominator = 2*tau*tau
        fraction_value = (-1*numerator)/denominator
        W[i,i] = np.exp(fraction_value)
        
    return W
```

```python
"""
Here I am making the function like where X will accepted as the form (m,n) where m = Total Examples and n = Total 
Features and Y is of the form (m,1). Query x should be of the form (1,n).
"""
def predict(X,Y,query_x,tau):
    
    #Adding (X0=1) in the first column to make our X matrix complete. It won't affect the net 
    #answer. It was just done so that when we do a dot product of X and theta (to find predicted Y), theta0 gets 
    #adjusted and we don't need to use for loop for that
    ones = np.ones((X.shape[0],1))
    X_ = np.concatenate((ones,X),axis = 1)
    
    #Adding 1 to query_x also as x0 as an extra parameter as done above for X matrix
    query_x_ones = np.array([
        [1]
    ])
    query_x = np.concatenate((query_x_ones,query_x),axis = 1)
    
    
    W = getW(query_x,X_,tau)
    
    
    #Now for getting theta first convert all of them into a matrix, so that instead of writing dot every time to do matrix 
    #multiplication, it will happen with * too in that matrix
    X_ = np.mat(X_)
    query_x = np.mat(query_x)
    W = np.mat(W)
    X_transpose = np.mat(np.transpose(X_))
    Y = np.mat(Y)

    
    #Here theta return will be of the form (n,1)
    theta = np.linalg.pinv(X_transpose*(W*X_))*(X_transpose*(W*Y))
        
    predicted_y = np.dot(query_x,theta) #It will get us the answer because query_x is of the form(1,n) and theta as(n,1).
    #Therefore their dot product will happen to be a scalar only as discussed before
    
    return predicted_y
```

### Making Prediction using the above value

```python
X = pd.read_csv('weightedX.csv').values
Y = pd.read_csv('weightedY.csv').values
```

```python
X.shape , Y.shape
```

```python
#Normalising the data
X = (X-X.mean())/X.std()
```

```python
#Now let's see how the plot looks like 
plt.style.use('seaborn')
plt.scatter(X,Y)
plt.show()
```

```python
#Now let's predict value of query point (X = 0.25).
X_query = np.array([
    [0.25]
])
tau = 1
predicted_Y = predict(X,Y,X_query,tau)
plt.style.use('seaborn')
plt.scatter(X,Y,color = 'blue')
print(X_query,predicted_Y)
plt.plot(X_query[0][0],predicted_Y[0][0],'ro')
```

```python
#Now let's see what happens if we increase and decrease the tau for the above example
tau = 0.1
predicted_Y = predict(X,Y,X_query,tau)
plt.style.use('seaborn')
plt.scatter(X,Y,color = 'blue')
print(X_query,predicted_Y)
plt.plot(X_query[0][0],predicted_Y[0][0],'ro')
plt.show()
```

```python
"""
Therefore clearly on reducing the tau, bandwidth decreases i.e contribution in loss for far away points for this
query point will decrease and more weigtage will be given to nearer points. Therefore it will be far away from a 
linear regression.
But if we increase this tau , then Bandwidth increases and more points which are farther away will contribute towards
loss, therefore on increasing tau it will more more nearer to normal linear regression
""" 
```

### Effect of TAU on Locally Weighted Regression 

```python
#Generated 20 points from -2 to 2 where each point is located at constant intervals of time 
#Now we will find the value of all these X and plot them
X_test = np.linspace(-2,2,20) 
print(X_test.shape)
print(X_test)
```

```python
def plot_prediction(tau):
    Y_test = []
    for x in X_test:
        x_query = np.array([
            x
        ]).reshape((1,-1))
        predictedY = predict(X,Y,x_query,tau)
        Y_test.append(predictedY[0][0])
    
    Y_test = np.array(Y_test).reshape((-1,1))
    plt.title('Bandwidth ' + str(tau))
    plt.scatter(X,Y,color = 'red')
    plt.scatter(X_test,Y_test,color='blue')
    plt.show()
```

```python
tau = [0.1,0.5,1,5]
for t in tau:
    plot_prediction(t)
```

```python
#Therefore clearly as we increase the tau , it moves more towards the normal linear regression as more farther points
#will contribute more towards the loss and hence it will affect it
```
