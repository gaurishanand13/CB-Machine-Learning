```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Visualising Y = (X-5)^2 as discussed in Notes too

```python
X = np.arange(10)
Y = (X-5)**2  ##Used Broadcasting to form the numpy array Y
print(X)
print(Y)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
#Now we need to find the minima of this function using gradient descent update rule --
#We already know and can clearly see that minima exists at X=5 , but let's confirm it too
```

```python
plt.plot(X,Y)
#Step1 )  Initialise X with some random Value. E.g - let it be 0

x = 0
learningRate = 0.1

#For now let's just take max of 50 steps
for i in range(50):
    
    #Let's plot the x,y after every updation on the graph
    plt.scatter(x,(x-5)**2)
    
    slope_at_x = 2*(x-5)  ##As derivate of f(x) = (x-5)^2 is 2*(x-5)
    
    if slope_at_x==0:
        break #As we have already reached the minima
        
    x = x - learningRate*slope_at_x
    
plt.show()
```

```python
#Let's print the values of x too with every updation
x = 0
learningRate = 0.1

for i in range(50):
    print(x,end = ' ')
    
    slope_at_x = 2*(x-5)
    
    if slope_at_x==0:
        break #As we have already reached the minima
        
    x = x - learningRate*slope_at_x
    
#Clearly we are reaching as close to 5 as possible in each step i.e moving towards minima
```

```python
#Now let's visualise when x is initialised with the value > 5 i.e in next half of convex. This time x will get reduce
#and try to reach the minima
plt.plot(X,Y)

x = 10
learningRate = 0.1

for i in range(50):
    #Let's plot the x after every updation
    plt.scatter(x,(x-5)**2)
    
    slope_at_x = 2*(x-5)
    
    if slope_at_x==0:
        break #As we have already reached the minima
        
    x = x - learningRate*slope_at_x
    
plt.show()
```
