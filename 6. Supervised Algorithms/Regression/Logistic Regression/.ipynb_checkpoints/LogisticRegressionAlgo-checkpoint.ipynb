### Data Preparation

```python
#Data 1 - Let these point belong to class 1
mean_01 = np.array([1,0.5])
cov_01 = np.array([
    [1,0.1],
    [0.1,1.2]
])

#Generating 500 points which are normally distributed in 2 directions with above mean and co-variance
data_1 = np.random.multivariate_normal(mean_01,cov_01,500)
print(data_1.shape)

print(data_1[:5])
```

```python
#Data 2 - Let these point belong to class 0
mean_02 = np.array([4,5])
cov_02 = np.array([
    [1.21,0.1],
    [0.1,1.3]
])
data_2 = np.random.multivariate_normal(mean_02,cov_02,500)
print(data_2.shape)

print(data_2[:5])
```

###### Visualizing the above points - 

```python
plt.scatter(data_1[:,0],data_1[:,1],marker = 'o',label='Class 1')
plt.scatter(data_2[:,0],data_2[:,1],marker = '^',label='Class 0')
plt.legend()
plt.show()
```

###### Creating Training and Test Set using the above points

```python
ones_array = np.ones([500]).reshape((500,-1))
print(ones_array.shape)
```

```python
data_1 = np.concatenate((data_1,ones_array),axis = 1)
print(data_1.shape)
print(data_1[:5])
```

```python
zeros_array = np.zeros([500]).reshape((500,-1))
data_2 = np.concatenate((data_2,zeros_array),axis = 1)
print(data_2.shape)
print(data_2[:5])
```

```python
final_data = np.concatenate((data_1,data_2),axis = 0)
print(final_data.shape)
```

```python
np.random.shuffle(final_data)
final_data[:10]
```

```python
#### Using Skikit to Generate Testing and Training Sets . We can generate Data without it too as done before
from sklearn.model_selection import train_test_split
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(final_data[:,:-1],final_data[:,-1],train_size=0.8)
print(X_TRAIN.shape , Y_TRAIN.shape)
print(X_TEST.shape , Y_TEST.shape)
```

### LOGISTIC ALGORITHM

```python
#First adding X0 = 0 as 1st column in the before data
X_TRAIN = np.concatenate((np.ones((X_TRAIN.shape[0],1)),X_TRAIN),axis = 1)
print(X_TRAIN.shape)
```

```python
print(X_TRAIN[:10])
```

```python
def sigmoid(z):
    return 1.0/(1.0 + np.exp(-1.0*z))


"""
X will be of the form (m,n+1) where n is the number of features and m is the number of examples and Theta will be of
the form (n+1,)
"""
def hypothesis(X,theta):
    return sigmoid(np.dot(X,theta))
```

<img style="transform: rotate(270deg);width=100px;" src='explaination1.jpeg'>

```python
"""
X will be of the form (m,n+1) where n is the number of features and m is the number of examples and Theta will be of
the form (n+1,) and Y will be of the form (m,)
"""
def error(X,Y,theta):
    m = X.shape[0]
    error = 0.0
    
    for i in range(m):
        hx = hypothesis(X[i].reshape((1,-1)),theta)
        error = error + (Y[i]*np.log2(hx[0]) + (1-Y[i])*np.log2(1-hx[0]))
        
    return -error/m
```

```python
"""
Here X will be of the form (m,n+1) , Y will be of the form (m,) and theta will be of the form (n+1,)
grads will be returned of the form (n+1,)
"""    

def get_gradients(X,Y,theta):
    
    gradients = np.zeros((X.shape[1],1))
    m = X.shape[0]
    
    for i in range(m):
        current_x = X[i].reshape((1,-1))
        hx = hypothesis(current_x,theta)
        value = (Y[i] - hx[0])*current_x.reshape((-1,1))
        gradients = gradients + value
        
        
    gradients = gradients.reshape((-1))
    return gradients/m

```

```python
#### Get Accuracy
def predict(X,theta):
    confidence = hypothesis(X.reshape((1,-1)),theta)
    if(confidence<0.5):
        return 0
    else:
        return 1

    
def get_acc(X_test,Y_test,theta):
    
    y_predicted = []
    
    for i in range(Y_test.shape[0]):
        p = predict(X_test[i],theta)
        y_predicted.append(p)
        
    y_predicted = np.array(y_predicted)
    
    return  float((y_predicted==Y_test).sum())/Y_test.shape[0]
```

<img style="transform: rotate(270deg);width=100px;" src='explaination2.jpeg'>

```python
#### GRADIENT DESCENT FUNCTION
def gradient_descent(X,Y,learning_rate=0.1,maximum_steps=1000):
    err = []
    accuracy = []
    
    theta = np.zeros((X.shape[1]))
    for i in range(maximum_steps):
        e = error(X,Y,theta)
        err.append(e)
        
        acc = get_acc(X,Y,theta)
        accuracy.append(acc)
        
        gradients = get_gradients(X,Y,theta)
        theta = theta + learning_rate*gradients
    
    return err,accuracy,theta
```

```python
err,accuracy,theta = gradient_descent(X_TRAIN,Y_TRAIN)
```

```python
err = np.array(err)
accuracy = np.array(accuracy)
print(err.shape)
print(accuracy.shape)
print(theta.shape)
```

```python
plt.plot(err,label = 'Error at every Iteration')
plt.plot(accuracy , label = 'Accuracy at every Iteration')
plt.legend()
plt.show()
```

Therefore clearly seeing the above graph , we can say that error kept on decreasing with every iteration and accuracy kept on increasing with every iteration

```python
print('Final Accuracy =',accuracy[-1]) ##Therefore through this algo we achieved an accuracy of 99.375 percent
```
