#### Data Visualisation of surface Plots

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

##### First Learning about meshgrid Function 

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7])

print(a.shape,a)
print(b.shape,b)
```

```python
a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
print(a.shape)
print(b.shape)
```

```python
plt.style.use('seaborn')
figure = plt.figure()
axes = figure.gca(projection='3d')
plt.show()
```

```python
def f(a,b):
    return a+b;

figure = plt.figure()
axes = figure.gca(projection='3d')
axes.plot_surface(a,b,f(a,b))
plt.show()
```

```python
figure = plt.figure()
axes = figure.gca(projection='3d')
axes.plot_surface(a,b,f(a,b),cmap='coolwarm')
plt.show()
```

```python
figure = plt.figure()
axes = figure.gca(projection='3d')
axes.plot_surface(a,b,f(a,b),cmap='rainbow')
plt.show()
```

```python
a = np.arange(-1,1,0.02)
print(a.shape)
```

```python
b = a
a,b = np.meshgrid(a,b)
print(a.shape)
print(b.shape)
```

```python
#Now let the function to be plotted --
def f(a,b):
    return a**2 + b**2

figure = plt.figure()
axes = figure.gca(projection='3d')
axes.plot_surface(a,b,f(a,b),cmap='rainbow')
plt.show()

##Therefore clearly when the f(x) is of the type a^2 + b^2 , then the 3D Plot will look like this -
```

#### Contour Plot for above - When the above Plot is divided by planes parallely to X,Y plane at some particular distance a contour plot is obtained

```python
figure = plt.figure()
axes = figure.gca(projection='3d')
axes.contour(a,b,f(a,b),cmap='rainbow')
plt.show()
```
