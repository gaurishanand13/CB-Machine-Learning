```python
import cv2  #Module of open cv
```

```python
myImage = cv2.imread('dogsImage.jpeg') #Reading the image using openCv
```

```python
#Now Image read using , OpenCV. It can be displayed using 2 Methods -
#1) Using Matplotlit
#) Using openCV
```

#### Displaying image using matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(myImage)
plt.show()
```

```python
#Clearly in above, channels were overlapped in wrong order (They were overlapped in opposite order i.e BGR).
#Now what we want is that, they should overlap in the original order i.e RGB. Therefore convert it then
finalImage = cv2.cvtColor(myImage,cv2.COLOR_BGR2RGB)#Converting the colour
plt.imshow(finalImage)
plt.show()
```

```python
#We can also display the image in black and white format like this-->
grayImage = cv2.imread('dogsImage.jpeg',cv2.IMREAD_GRAYSCALE)
finalGrayImage = cv2.cvtColor(grayImage,cv2.COLOR_BGR2RGB)
plt.imshow(finalGrayImage)
plt.show()
```

#### Displaying Image using openCV

```python
#And if we are showing image using openCV, we don't need to convert it in the form of RGB format. It prints by default
#in RBG format
cv2.imshow('Dog Image',myImage) #String 'Dog Image' is the title given to the image window displayed

#Note cv2.waitKey(x) waits for x milli-seconds for a key to be pressed. If key is not pressed before x seconds , then
#it returns -1 otherwise it returns the 8digit - ascii value of the the key pressed from the keyboard. We can even see
#it

#Note if x = 0, then it waits for infinite second till the user presses any key. 
#Therefore only after a key is pressed, then only it will execute the code after this line
cv2.waitKey(0) 


cv2.destroyAllWindows()#This destroys all the new windows created by the openCV, Maybe to show image or video capture
#Therefore in this case , it will close the image window
```

```python
cv2.imshow('Dog Image',myImage)
x = cv2.waitKey(10000) #It will wait for maximum 10 seconds for us unitl we press a key otherwise returns -1
print(x)
cv2.destroyAllWindows()

#Clearly Here no key was pressed,  so we found got output as -1
```

```python
cv2.imshow('Dog Image',myImage)
x = cv2.waitKey(10000) ## x will have the ascii code of the character pressed
cv2.destroyAllWindows()

#Since here the key pressed was 'd', Therefore it's ascii code is printed
```

```python
#We can even confirm the ascii code of above like this
#Ord gives the ascii code of the character pressed
asciiValue  = ord('d')
print(asciiValue)
```

```python
#Displaying the black and white image , don't need to convert the channels here
grayImage = cv2.imread('dogsImage.jpeg',cv2.IMREAD_GRAYSCALE)
cv2.imshow('Dog Image',grayImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
